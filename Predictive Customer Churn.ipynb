## **Project Domain**

Churn prediction means detecting which customers are likely to leave a service or cancel a service subscription. This is an important prediction for many businesses because acquiring new customers is often more expensive than retaining existing customers. 

Every time a customer leaves, it is arguably a lost investment. Both time and effort must be channeled to replace them. Being able to predict when a customer will leave, and offering them incentives to stay, can save the company money.

Different customers have different traits. It can be very time-consuming to analyze whether a customer will churn or not. Therefore, it is very important for a business to be able to automate in predicting which customers will churn or not. Later, the results of this prediction can be used to create attractive incentives/offers.

Referensi : [Customer Churn](https://www.avaus.com/blog/predicting-customer-churn/#:~:text=Predicting%20Customer%20Churn,more%20than%20retaining%20existing%20ones.)

## **Business Understanding**

### **Problem Statements**

1. A high customer churn rate can be an indication of deficiencies in service or inadequate promotion. Thus, telecom companies need to understand what factors drive customers to unsubscribe.

2. Identifying patterns and trends associated with customer churn can help companies to anticipate and prevent future customer loss. However, success in this regard depends on the ability to effectively manage and analyze customer data.

### **Goals**

1. Identify the key factors that contribute to customer churn. By understanding these factors, companies can take appropriate actions to improve customer retention and reduce churn rates.

2. Develop a predictive model that can predict customer churn with high accuracy. Thus, companies can take preventive actions quickly to retain customers who may be at risk of unsubscribing.

**Take a look about correlation in target and other features**

To know the correlation between categorical data, we can use chisquare method.

```python
from scipy.stats import chi2_contingency
# check correlation between features and target
def chi2_test(feature_name, target):

    contingency_table = pd.crosstab(dataset[feature_name], dataset[target])

    chi2, p, dof, expected = chi2_contingency(contingency_table)

    alpha = 0.05

    significance_level = 'Significant' if p < alpha else 'Not Significant'


    summary = pd.DataFrame({
        'Feature': [feature_name],
        'Chi-Square Statistic': [chi2],
        'P-Value': [p],
        'Alpha Value': [alpha],
        'Significance Level': [significance_level]
    })

    return summary
```

```python
significance_summary = pd.DataFrame({
        'Feature': [],
        'Chi-Square Statistic': [],
        'P-Value': [],
        'Alpha Value': [],
        'Significance Level': []
})

for col in dataset.columns:
    if col == 'Churn':
        pass
    elif col == 'tenure':
        pass
    elif col == 'MonthlyCharges':
        pass
    else:
        result = chi2_test(col, 'Churn')
        significance_summary = pd.concat([significance_summary, result])
```

## **Data Preparation**

```python
main_data = dataset.drop(columns=['gender', 'PhoneService'])
```

```python
main_data['total_services'] = 0

services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
            'TechSupport', 'StreamingTV', 'StreamingMovies']

def service_score(col):
    for index, row in main_data.iterrows():
        if row[col] == 'Yes':
            main_data.at[index, 'total_services'] += 1

# Loop through each service column and call service_score function
for col in services:
    service_score(col)
```

**Split the data**

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Split the data and make sure Stratified 
X = main_data
y = main_data['Churn']
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)

for train_index, test_index in sss.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index] 
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
X_train, X_test = X_train.drop(columns='Churn'),  X_test.drop(columns='Churn')
```

## **Data Preprocessing**

```python
num_features = X_train.describe().columns.to_list()
cat_features = X_train.columns[~X_train.columns.isin(X_train.describe().columns)].to_list()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline

# One hot encoding for categorical data
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder())
])

# Normalization data
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Combine transformers for categorical and numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, cat_features),
        ('num', numerical_transformer, num_features)
    ])

# Fit the preprocessor to your training data
preprocessor.fit(X_train)

# Transform 
X_train_encoded = preprocessor.transform(X_train)
X_test_encoded = preprocessor.transform(X_test)
```

```python
le = LabelEncoder()

le.fit(y_train)

y_train_encoded = le.transform(y_train)
y_test_encoded = le.transform(y_test)
```

## **Model Development**

Because this is the classification problem. I will used some machine learning algorithm that good when handling classification problem. This include : RandomForest, KNN Classifier, Logistic Regression, AdaBoost.

```python
# To store model results
model_results = {}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
```

```python
# Create a Random Forest classifier instance
rf = RandomForestClassifier(n_estimators=30, max_depth=8, random_state=42)

rf.fit(X_train_encoded, y_train_encoded)

y_test_predicted_rf = rf.predict(X_test_encoded)

model_results['rf'] = (y_test_encoded, y_test_predicted_rf)
print(classification_report(y_test_encoded, y_test_predicted_rf))
```

```python
knn = KNeighborsClassifier(n_neighbors=250, weights='uniform') 

knn.fit(X_train_encoded, y_train_encoded)

y_test_predicted_knn = knn.predict(X_test_encoded)

model_results['knn'] = (y_test_encoded, y_test_predicted_knn)
print(classification_report(y_test_encoded, y_test_predicted_knn))
```

```python
lr = LogisticRegression(random_state=42)

lr.fit(X_train_encoded, y_train_encoded)

y_test_predicted_lr = lr.predict(X_test_encoded)

model_results['lr'] = (y_test_encoded, y_test_predicted_lr)
print(classification_report(y_test_encoded, y_test_predicted_lr))
```

```python
ada = AdaBoostClassifier(n_estimators=20, random_state=42)

ada.fit(X_train_encoded, y_train_encoded)

y_test_predicted_ada = ada.predict(X_test_encoded)

model_results['ada'] = (y_test_encoded, y_test_predicted_ada)
print(classification_report(y_test_encoded, y_test_predicted_ada))
```

## **Evaluation**

```python
# Initialize lists to store metrics
model_names = []
precisions = []
recalls = []
f1_scores = []
accuracies = []

# Iterate through each model
for model_name, (y_true, y_pred) in model_results.items():
    # Calculate precision, recall, F1 score, accuracy
    accuracy = accuracy_score(y_true, y_pred)
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_true, y_pred, labels=[1], average=None)
    
    # Append metrics to lists
    model_names.append(model_name)
    precisions.append(round(precision[0], 2))
    recalls.append(round(recall[0], 2))
    f1_scores.append(round(f1_score[0], 2))
    accuracies.append(round(accuracy, 2))

# Create a DataFrame to store the metrics
summary_table = pd.DataFrame({
    'Model': model_names,
    'Precision': precisions,
    'Recall': recalls,
    'F1 Score': f1_scores,
    'Accuracies': accuracies
})

# Display the summary table
summary_table
```

Above table is summary about the performance machine learning models when predict customer who would Churn. The best performance models is logistic regression. the precision is close to 70, and the accurcay is above 80. Also F1 Score for logistic regression model is the highest among all models. So, the logistic regression is the best model for predict customers who would Churn.
